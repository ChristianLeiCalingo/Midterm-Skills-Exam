**Concatenating the 2 datasets**

**Let's check the values in one of the columns**

```python
data['native-country'].unique() #checking for the unique values of country
```

**As you can see, we have a ? value for the native country, other columns might have ? values as well, we can change those values to Others so that other data will not be wasted**

```python
replace_nan = {"?" : "Others"}

data.replace(replace_nan, inplace = True) #changing the values of the unidentified
data.fillna('Others', inplace = True)
data
```

**Now, we wil check if we have replaced it properly**

```python
data.isna().any()
```

**As you can see, each column returned False, therefore, we don't have NaN values anymore**


---

**Now, we are going to set the country as its index**

```python
data.rename(columns = {'native-country' : 'country'}, inplace = True)
data.set_index(data['country'],inplace = True) #setting the country as index
data.drop('country', axis = 1, inplace = True)
data
```


---
**At the income column, we can see that we have 4 unique values:**
* (<=50K)
* (<= 50K.)
* (>50K)
* (>50K.)

**We are going to remove the decimal after the 'K'**



```python
data['income'].unique() #checking for the unique values of income
```

```python
row_change = {'<=50K.' : '<=50K', #changing the value because of the unecessary .
              '>50K.' : '>50K'}

data = data.replace({'income' : row_change})
data
```

**Let's plot the Marital Status, Relationship, Sex, Race, and Education of our Population to identify their quantity.**

**Since we won't be needing them on our Analysis, we will be dropping them except for the Education**

**Here, we have the average capital-gain and capital-loss of each country**

```python
data_cap = data[['capital-gain', 'capital-loss', 'hours-per-week']]
data_cap = data_cap.groupby('country').mean()
data_cap.sort_values(by = 'country', ascending = True, inplace = True)
data_cap
```

**We can graph it using subplots**

```python
fig, ax = plt.subplots(3)

data_cap['capital-gain'].sort_values(ascending = True).plot(ax = ax[0], kind = 'barh', figsize = (10,20), grid = True,
                                                            title = 'Mean Capital Gain for each Country', color = ('cadetblue', 'steelblue'))

data_cap['capital-loss'].sort_values(ascending = True).plot(ax = ax[1], kind = 'barh', figsize = (10,20), grid = True,
                                                            title = 'Mean Capital Loss for each Country', color = ('darkolivegreen', 'olivedrab'))

data_cap['hours-per-week'].sort_values(ascending = True).plot(ax = ax[2], kind = 'barh', figsize = (10,20), grid = True,
                                                            title = 'Working Hours Per Week of each Country', color = ('sandybrown', 'saddlebrown'))
fig.tight_layout()
```

**Looking at the Graph Above, Inda have the Highest Average Gain followed by Japan and Taiwan**

**For the capital loss, Holand-Netherlands take the top spot followed by Hongkong and Cambodia**

**In terms of Hours-Per-Week, Greece is the first followed by Thailand and Iran**


---



**Let's check if there is a correlation between the Capital-Gain, Capital-Loss for each country and their Hours Per Week in working**

**Using sns.heatmap(), we can check their correlation**

```python
sns.heatmap(data_cap.corr(), annot = True)
```

* **Correlation coeffictient between the Hours Per Week and Capital Gain is 0.33. Therefore, we have a Weak Positive Correlation**

* **Correlation coeffictient between the Hours Per Week and Capital Loss is 0.0029. Therefore, we have a Very Weak Postive Correlation**

* **Correlation coeffictient between the Capital Gain and Capital Gain is -0.12. Therefore, we have a Very Weak NegativeCorrelation**
